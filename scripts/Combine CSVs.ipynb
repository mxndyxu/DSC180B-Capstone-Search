```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
projectsDF_1920 = pd.read_csv("projectsDF_1920.csv", index_col=0)
projectsDF_2021 = pd.read_csv("projectsDF_2021.csv", index_col=0)
projectsDF_2122 = pd.read_csv("projectsDF_2122.csv", index_col=0)
projectsDF_2223 = pd.read_csv("projectsDF_2223.csv", index_col=0)
```

```python
projectsDF_1920["Year"] = [2020 for i in range(projectsDF_1920.shape[0])]
projectsDF_2021["Year"] = [2021 for i in range(projectsDF_2021.shape[0])]
projectsDF_2122["Year"] = [2022 for i in range(projectsDF_2122.shape[0])]
projectsDF_2223["Year"] = [2023 for i in range(projectsDF_2223.shape[0])]
```

```python
comb_projectsDF = pd.concat([projectsDF_1920, projectsDF_2021, projectsDF_2122, projectsDF_2223], ignore_index=True, axis=0)
```

```python
comb_projectsDF.to_csv("comb_projectsDF.csv")
```
